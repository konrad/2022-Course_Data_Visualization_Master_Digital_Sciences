```python
import urllib.request
import pandas as pd
import seaborn as sns

# Downlaoding the data as saving it as local file
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
csv_file = "penguins.csv"
urllib.request.urlretrieve(url, csv_file)
```

```python
# Load the data into a pandas DataFrame
penguins = pd.read_csv(csv_file)
```

```python
penguins
```

### Using pandas simple in-built plotting capabilities

```python
penguins["body_mass_g"].describe()
```

```python
# Increasing the granularity by highering the number of bins
penguins["body_mass_g"].hist(bins=20)
```

# Scatterplot

In the following we will encode an increasing amount of attributes (numberical and categorial) by different aesthetics.

- positional x-axis
- positional y-axis
- area size
- color
- shape

```python
# Plotting two numerical atttibutes - one on the x axis, one on the y-axis

sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
```

```python
# Now encode a third numerical attribue in the area size of the dots
# (this could now be called a bubble lot)

sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", size="body_mass_g")
```

```python
# Now enconding a fouth atttibute in the color value of the plot
sns.scatterplot(data=penguins,
                x="bill_length_mm",
                y="bill_depth_mm",
                size="body_mass_g",
                hue="flipper_length_mm")
```

```python
# Now plotting two numerical attributes in the position 
# and one categorical attribut in the color.

sns.scatterplot(data=penguins,
                x="bill_length_mm",
                y="bill_depth_mm",
                hue="species")
```

```python
# Now encoding numerial attributes and one (another) categorical attribut in the color

sns.scatterplot(data=penguins,
                x="bill_length_mm",
                y="bill_depth_mm",
                hue="island")
```

```python
# Now plotting numerial attributes and on categorical attribut in the shape of the dots
# (harder to distiguis than color)

sns.scatterplot(data=penguins,
                x="bill_length_mm",
                y="bill_depth_mm",
                style="island")
```

```python
# Now plotting two numerial attributes in position, one categorial 
# attribute in color and one categorial attribute by shape

sns.scatterplot(data=penguins,
                x="bill_length_mm",
                y="bill_depth_mm",
                style="island",
                hue="sex")
```

```python
# Encoding two numerical values in the positions,
# one numerical in the size and two categorial 
# attributes in the color and in the shape.

# Warning - not necessarily a good idea

sns.scatterplot(data=penguins,
                x="bill_length_mm",
                y="bill_depth_mm",
                size="body_mass_g",
                style="island",
                hue="sex")
```

```python
# Not the categorial attribue "island" is encoded by shape and color; making
# recognizing the classes easier

sns.scatterplot(data=penguins,
                x="bill_length_mm",
                y="bill_depth_mm",
                size="body_mass_g",
                style="island",
                hue="island")
```

```python
# bad!
# Here the numerical attribute "body_mass_g" is encoded in the position on
# the y axis and the area size.
sns.scatterplot(data=penguins,
                x="bill_length_mm",
                y="body_mass_g",
                size="body_mass_g",
                style="island",
                hue="island")
```
